8. Kolom Hospital memiliki atribut kategorikal jenis nominal. Kolom ini terdiri atas data nama rumah sakit di mana pasien terdaftar.
9. Kolom Insurance Provider memiliki atribut kategorikal jenis nominal. Kolom ini terdiri atas data nama penyedia asuransi yang pasien gunakan. Nilainya dapat berupa salah satu dari opsi: "Aetna," "Blue Cross," "Cigna," "UnitedHealthcare," dan "Medicare."
10. Kolom Billing Amount memiliki atribut kuantitatif jenis diskrit. Kolom ini terdiri atas data jumlah biaya yang harus dikeluarkan pasien atas pelayanan kesehatan yang sudah didapatkan. Besar tagihan yang harus dibayar pasien berkisar antara 1000.1808370217516 sampai dengan 49995.90228322116
11. Kolom Room Number memiliki atribut kategorikal jenis nominal. Kolom ini berisi data nomor kamar di mana pasien tinggal selama perawatan.
12. Kolom Admission Type memiliki atribut kategorikal jenis nominal. Kolom ini berisi data jenis pendaftaran yang dilakukan pasien berdasarkan kondisi. Nilainya dapat berupa "Emergency," "Elective," atau "Urgent."
13. Kolom Discharge Date memiliki atribut kuantitatif jenis kontinu. Nilainya merupakan tanggal kapan pasien meninggalkan rumah sakit.
14. Kolom Medication memiliki atribut kategorikal jenis nominal. Kolom ini berisi data obat yang diterima pasien. Nilainya dapat berupa "Aspirin," "Ibuprofen," "Penicillin," "Paracetamol," dan "Lipitor."
15. Kolom Test Results memiliki atribut kategorikal jenis nominal. Kolom ini berisi data luaran pasien saat menjalani tes kesehatan. Nilai yang mungkin adalah  "Normal," "Abnormal," atau "Inconclusive."

***Atribut 'Age'***

```python


# Rata-rata 
rata_age = data['Age'].mean()
print(f"Rata-rata dari data umur adalah {rata_age}")

# Standar Deviasi
std_age = data['Age'].std()
print(f"Standar Deviasi dari data umur adalah {std_age}")

# Modus
modus_age = data['Age'].value_counts().idxmax()
print(f"Modus dari data umur adalah{modus_age}")

# Median
median_age = data['Age'].median()
print(f"Median dari data umur adalah {median_age}")

# Persentil
# Persentil(10%)
_01_age = data['Age'].quantile(0.10)
print(f"Persentil 10% dari data umur adalah {_01_age}")
# Persentil(25%)
_025_age = data['Age'].quantile(0.25)
print(f"Persentil 25% dari data umur adalah {_025_age}")
# Persentil(50%)
_050_age = data['Age'].quantile(0.50)
print(f"Persentil 50% dari data umur adalah {_050_age}")
# Persentil(25%)
_075_age = data['Age'].quantile(0.75)
print(f"Persentil 75% dari data umur adalah {_075_age}")
# Persentil(90%)
_090_age = data['Age'].quantile(0.90)
print(f"Persentil 90% dari data umur adalah {_090_age}")

# Nilai Maksimum
max_age = data['Age'].max()
print(f"Nilai maksimum dari data umur adalah {max_age} tahun")

# Nilai Minimum
min_age = data['Age'].min()
print(f"Nilai minimum dari data umur adalah {min_age} tahun")

# Distribusi Frekuensi
frek_age = pd.cut(data['Age'],bins=[15,30,45,60,75,90,]).value_counts().sort_index()
print('_________________')
print(frek_age)
print('_________________')
```

***Data jumlah pasien sejalan dengan waktu***

```python
from datetime import datetime

df = df.sort_values(['Date of Admission'], ascending=1) # mengurutkan data berdasarkan waktu.

df['Date of Admission'] = pd.to_datetime(df['Date of Admission']) # mengubah jenis data waktu dari string menjadi datetime

df['year'] = pd.DatetimeIndex(df['Date of Admission']).year # mengambil data tahun setiap pasien memasuki rumah sakit

df.groupby('year').size().plot(kind='line', color=[ 'red'], figsize=(16,8),
                              title='Data jumlah pasien pada setiap penyakit sejalan dengan waktu')
```

Insight : Awalnya jumlah pasien terlihat sedikit di tahun 2018, lalu melonjak tinggi di 2019. Alasannya, karena data pasien mulai dicatat akhir tahun 2018 (bulan oktober). Lalu, dari tahun 2019, jumlah pasien terlihat terus bertambah di tahun 2020 dan 2021. Baru kemudian terlihat menurun di 2022 dan 2023.

***Data jumlah pasien setiap penyakit sejalan dengan waktu***

### Data cleansing - Bonus

Untuk mengecek kekotoran data, kami menggunakan syntax “.info()”. Hasil yang ditampilkan oleh syntax ini adalah berupa jumlah data yang error atau data yang tidak terdefinisi nilainya.

```python
dataframe = pd.read_csv("healthcare_dataset.csv")

print("banyak data =" ,len(dataframe)) # mengecek banyak data
print("")
print("")
dataframe.info() # mengecek data terdefinisi
```

Agar data yang dianalisis akurat, Data tersebut harus memiliki tingkat kekotoran data seminimal mungkin.Terlihat dari gambar di atas bahwa dari jumlah data dan informasi tersebut, semua atribut tidak memiliki error atau nilai yang tidak terdefinisi.
